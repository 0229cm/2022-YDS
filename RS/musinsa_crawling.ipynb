```python
seed = np.random.randint(100)
np.random.seed(seed)
a = np.random.randint(3)
time.sleep(a)
```

```python
# brands = list()
# items = list()
# prices = list()
# stars = list()
# likes = list()
datas = list()
for page in range(3,12):
    
    for i in range(90):
        # brand
        brand = html.select_one('.list-box.box').select('.li_box')[i].find('p',"item_title").find('a')

        if brand is None:
            datas.append(0)
        else:
            datas.append(brand.text)    
        # item
        item = html.select_one('.list-box.box').select('.li_box')[i].find('p',"list_info").find('a').attrs['title']
        datas.append(item)
        # 가격
        price = html.select_one('.list-box.box').select('.li_box')[i].find('p',"price").text.replace(' ','')
        datas.append(price)
        # 리뷰 평점
        star = html.select_one('.list-box.box').select('.li_box')[i].find('div',"article_info").find('span',"count")
            # 평점 데이터가 없는경우
        if star is None:
            datas.append(0)
        else:
            datas.append(star.text)
        # 좋아요 갯수
        like = html.select_one('.list-box.box').select('.li_box')[i].find('div',"article_info").find('span',{'name':'count'})
                # 좋아요 데이터가 없는경우
        if like is None:
            datas.append(0)
        else:
            datas.append(like.text)
    
    # 차단막기
    time.sleep(a)
    
    #다음페이지 이동
    # click()으로 하니까 계속 에러남.. send_keys(Keys.ENTER)
    driver.find_elements_by_css_selector("#goodsRankForm > div.right_contents.hover_box > div.boxed-list-wrapper.rank-shop > div.pagination-box.box > div > div > a")[page].send_keys(Keys.ENTER)
#     datas.append(like)
datas
```
